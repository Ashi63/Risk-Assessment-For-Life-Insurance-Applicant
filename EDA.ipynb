```python
# checking for missing values in all the data columns
def checking_missing_values(data):
  """This function gives details of missing values and return a dataframe
     with two columns, missing values count and the percentage of missing values.
     -> Takes original dataframe as parameter
     <- Returns dataframe with details of missing values
  """
  missing_values_details = pd.DataFrame(data.isnull().sum(),columns=['missing_values']) 
  missing_values_details['missing_values_%']=data.isnull().sum()/len(data)*100
  return missing_values_details.sort_values(ascending=False,by='missing_values_%')



# removing columns with more than 50% missing values in all the data columns
def remove_missing_value_columns(data,per_missing_value):
  """This method removes column with more than 40 % missing values.
     <- Return data frame with colums removed with more than 40% percent missing values
  """
  missing_values_col_list = missing_values_details[missing_values_details['missing_values_%']>per_missing_value].index.values.tolist()
  total_missing_values = missing_values_details[missing_values_details['missing_values_%']!=0].index.values.tolist()
  print("Total Missing Values Columns List : ",total_missing_values)
  print("Total Number of columns with missing values: ",len(total_missing_values))
  print("Number Columns with missing values more than 40%: ",len(missing_values_col_list))
  print("Shape of Data Before droping column: ",data.shape)
  data1 = data.drop(columns=missing_values_col_list,axis=1)
  print("Shape of Data After droping column: ",data1.shape)
  print("No. of columns droped: ",data.shape[1]-data1.shape[1])
  return data1

missing_values_details = checking_missing_values(data)
print("Missing Value details: \n",missing_values_details.head(15))

data = remove_missing_value_columns(data,40)
data


# checking missing values in data columns which are left after
# removing the columns with more than 50% missing values
data2 = checking_missing_values(data)

# getting list of columns with missing values
remaining_colums_with_missing_values = data2[data2['missing_values']!=0].index.values.tolist()
print("\n")
print("List of columns with missing values less than 50% : ",remaining_colums_with_missing_values)
print("Total Number of Columns with missing value less than 50% :",len(remaining_colums_with_missing_values))
print('\n')

# creating a dataframe with missing value columns for analysis.
data_x= data[remaining_colums_with_missing_values]

# checking the details of missing values present
print("Details of missing values:\n",checking_missing_values(data_x))
print('\n')

# calculating mean,median for all our columns with missing values 
mean_Family_Hist_4 = data_x['Family_Hist_4'].mean()
median_Family_Hist_4 = data_x['Family_Hist_4'].median()

mean_Employment_Info_6 = data_x['Employment_Info_6'].mean()
median_Employment_Info_6 = data_x['Employment_Info_6'].median()

mean_Medical_History_1 = data_x['Medical_History_1'].mean()
median_Medical_History_1 = data_x['Medical_History_1'].median()

mean_Employment_Info_4 = data_x['Employment_Info_4'].mean()
median_Employment_Info_4 = data_x['Employment_Info_4'].median()

mean_Employment_Info_1 = data_x['Employment_Info_1'].mean()
median_Employment_Info_1 = data_x['Employment_Info_1'].median()

# filling missing values with mean, and median
data_x['mean_Family_Hist_4']=data_x['Family_Hist_4'].fillna(mean_Family_Hist_4)
data_x['median_Family_Hist_4']=data_x['Family_Hist_4'].fillna(median_Family_Hist_4)

data_x['mean_Employment_Info_6']=data_x['Employment_Info_6'].fillna(mean_Employment_Info_6)
data_x['median_Employment_Info_6']=data_x['Employment_Info_6'].fillna(median_Employment_Info_6)

data_x['mean_Medical_History_1']=data_x['Medical_History_1'].fillna(mean_Medical_History_1)
data_x['median_Medical_History_1']=data_x['Medical_History_1'].fillna(median_Medical_History_1)

data_x['mean_Employment_Info_4']=data_x['Employment_Info_4'].fillna(mean_Employment_Info_4)
data_x['median_Employment_Info_4']=data_x['Employment_Info_4'].fillna(median_Employment_Info_4)


data_x['mean_Employment_Info_1']=data_x['Employment_Info_1'].fillna(mean_Employment_Info_1)
data_x['median_Employment_Info_1']=data_x['Employment_Info_1'].fillna(median_Employment_Info_1)


# plotting and checking the distributions of columns when normal, when filled with mean, when filled with median
fig,axes = plt.subplots(nrows=5,ncols=3,figsize=(12,18))

sns.distplot(data_x['Family_Hist_4'],ax=axes[0,0],label='Normal',color='Red')
sns.distplot(data_x['mean_Family_Hist_4'],ax=axes[0,1],label='Mean',color='Blue')
sns.distplot(data_x['median_Family_Hist_4'],ax=axes[0,2],label='Median',color='Green')


sns.distplot(data_x['Employment_Info_6'],ax=axes[1,0],color='Red')
sns.distplot(data_x['mean_Employment_Info_6'],ax=axes[1,1],color='Blue')
sns.distplot(data_x['median_Employment_Info_6'],ax=axes[1,2],color='Green')

sns.distplot(data_x['Medical_History_1'],ax=axes[2,0],color='Red')
sns.distplot(data_x['mean_Medical_History_1'],ax=axes[2,1],color='Blue')
sns.distplot(data_x['median_Medical_History_1'],ax=axes[2,2],color='Green')

sns.distplot(data_x['Employment_Info_4'],ax=axes[3,0],color='Red')
sns.distplot(data_x['mean_Employment_Info_4'],ax=axes[3,1],color='Blue')
sns.distplot(data_x['median_Employment_Info_4'],ax=axes[3,2],color='Green')


sns.distplot(data_x['Employment_Info_1'],ax=axes[4,0],color='Red')
sns.distplot(data_x['mean_Employment_Info_1'],ax=axes[4,1],color='Blue')
sns.distplot(data_x['median_Employment_Info_1'],ax=axes[4,2],color='Green')


```
